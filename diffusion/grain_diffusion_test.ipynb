```python
import matplotlib.pyplot as plt 
import numpy as np 
from scipy.special import erfc
```

```python
# значение коэффициентов
D_balk = 5e-18
D_gb = D_balk * 1e4
delta = 1 * 1e-9
C_0 = 1e22
x = np.linspace(-10, 10, 100) * 1e-8
y = np.linspace(0, 10, 100) * 1e-9
time = 100 

C = C_0 * np.exp((y * np.power(D_balk, 1/4)) / (np.power(np.pi * time, 1/4) * np.power(delta * D_gb, 1/2))) * erfc(x / (2 * np.power(D_balk * time, 1/2)))

plt.plot(y, C)
#plt.yscale('log')
```
